```python
'''
Visualize small 4d unsafe set.
'''

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
import scipy.io
mat = scipy.io.loadmat('small-4d-unsafe-set.mat')
matGrid = mat["data"][:,:,:,:,-1]
```

```python
matGrid.shape
```

```python
p = np.pi
yaws = [0, 3, 6, 9, 12]
```

```python
#def ObstaclesFromMatLab():
fig, axs = plt.subplots(5,1)
print(matGrid.shape)
fig = plt.figure()
count = 1
for selected in yaws: 
    grid = []
    for x in range(len(matGrid)):
        for y in range(len(matGrid[0])):
          for yaw in range(len(matGrid[0][0])):
              for z in range(len(matGrid[0][0][0])):
                  if matGrid[x][y][yaw][z] <= 0:
                    if yaw == selected:
                        grid.append((x, y,z))
    x_path = [Node[0] for Node in grid]
    y_path = [Node[1] for Node in grid]
    z_path = [Node[2] for Node in grid]

    #ax = fig.add_subplot(5,1,count, projection='3d')
    ax = plt.axes(projection='3d')
    #ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
    ax.scatter(x_path,y_path,z_path)
    count+=1
    # ax.scatter(gridx,gridy,gridz)
    plt.xlabel('x')
    plt.ylabel('y')

    plt.show()
```
